# Getting to know matplotlib

## Making simple plots

### Simple line

```python
import matplotlib.pyplot as plt

# Got to have something to plot...
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# Creating the line chart with time on x-axis and gdp on y-axis
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# Chart title
plt.title('Nominal GDP')

# Label indicating the units for the y-axis
plt.ylabel('Billions of USD')

# Label for the x-axis
plt.xlabel('Year')

# Display the plot
plt.show()
```

### Bar charts

These are good for looking at quantities that vary among discrete sets of items.

```python
# Things to plot
movies = ['Annie Hall', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]

# Default bar size is 0.8, so ading  to left coordinate to center each bar
xs = [i + 0.1 for i, _ in enumerate(movies)]

# Plot bar with the xs x-coordinate and the number of oscars as the heights
plt.bar(xs, num_oscars)

# Give a title
plt.title("Some Movies and Their Awards")

# Give the y-axis a label
plt.ylabel("# of Academy Awards Won")

# Put the movie titles as the labels for each bar on the x-axis
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

# Display it
plt.show()
```

Bar charts are also used for histomgrams so we can visually see the distribution of values.

```python
from collections import Counter

# Stuff to plot
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade // 10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar([x - 4 for x in histogram.keys()], # shift each bar to the left by 4 units
       histogram.values(), # give each bar the correct height
       8) # bar width

# x-axis from -5 to 105 and y-axis from 0 to 5
plt.axis([-5, 105, 0, 5])

# x-axis labels at 0, 10, ..., 100
plt.xticks([10 * i for i in range(11)])

plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

### Line Charts

```python
# Stuff to plot
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# Make multiple calls to plt.plot to show multiple series on same chart
plt.plot(xs, variance, 'g-', label='variance') # green solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # red dot-dashed line
plt.plot(xs, total_error, 'b:', label='total error') # blue dotted line

# Create a legend since we have multiple series on same chart
# loc=9 means "top-center"
plt.legend(loc=9)
plt.xlabel("model complexity")
plt.ylabel("The Bias-Variance Tradeoff")
plt.show()
```

### Scatterplots

Good for visualizing relationships between two paired data sets, like number of friends and minutes spent on some website.

```python
# The data sets
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

# The scatter plot
plt.scatter(friends, minutes)

# Label each point on the plot
for label, friend_count,  minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                xy = (friend_count, minute_count), #put labels with their points
                xytext = (5, -5), # but make them slightly offset
                textcoords='offset points')
    
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of firends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```
